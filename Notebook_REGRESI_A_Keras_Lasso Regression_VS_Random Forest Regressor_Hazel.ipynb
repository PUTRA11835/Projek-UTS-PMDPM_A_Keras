```python
print("Sebelum pengecekan data duplikat, ", df_properti2.shape)
df_properti3=df_properti2.drop_duplicates(keep='last')
print("Setelah pengecekan data duplikat, ", df_properti3.shape)
```

```python
from sklearn.model_selection import train_test_split

x_regress = df_properti3.drop('price', axis=1)
y_regress = df_properti3.price

x_train_price, x_test_price, y_train_price, y_test_price = train_test_split(x_regress, y_regress, test_size=0.25, random_state=40)
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer

kolom_kategori=['hasyard', 'haspool', 'isnewbuilt', 'hasstormprotector', 'hasstorageroom']

transform = make_column_transformer(
    (OneHotEncoder(), kolom_kategori), remainder='passthrough'
)
```

```python
x_train_enc=transform.fit_transform(x_train_price)
x_test_enc=transform.fit_transform(x_test_price)

df_train_enc=pd.DataFrame(x_train_enc, columns=transform.get_feature_names_out())
df_test_enc=pd.DataFrame(x_test_enc, columns=transform.get_feature_names_out())

df_train_enc.head(10)
df_test_enc.head(10)
```

```python
from sklearn.linear_model import Lasso
from sklearn.model_selection import GridSearchCV
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.feature_selection import  SelectKBest, SelectPercentile, f_regression
from sklearn.metrics import mean_absolute_error, mean_squared_error

pipe_Lasso = Pipeline(steps=[
            ('scale', StandardScaler()),
            ('feature_selection', SelectKBest(score_func=f_regression)),
            ('reg', Lasso(max_iter=1000))            
            ])
pipe_Lasso2 = Pipeline(steps=[
            ('scale', StandardScaler()),
            ('feature_selection', SelectPercentile(score_func=f_regression)),
            ('reg', Lasso(max_iter=1000))            
            ])
pipe_Lasso3 = Pipeline(steps=[
            ('scale', MinMaxScaler()),
            ('feature_selection', SelectKBest(score_func=f_regression)),
            ('reg', Lasso(max_iter=1000))            
            ])
pipe_Lasso4 = Pipeline(steps=[
            ('scale', MinMaxScaler()),
            ('feature_selection', SelectPercentile(score_func=f_regression)),
            ('reg', Lasso(max_iter=1000))            
            ])

param_grid_Lasso = {
    'reg__alpha': [0.01,0.1,1,10,100],
    'feature_selection__k': np.arange(1, 20)
}
param_grid_Lasso2 = {
    'reg__alpha': [0.01,0.1,1,10,100],
    'feature_selection__percentile': np.arange(1, 20)
}
param_grid_Lasso3 = {
    'reg__alpha': [0.01,0.1,1,10,100],
    'feature_selection__k': np.arange(1, 20)
}
param_grid_Lasso4 = {
    'reg__alpha': [0.01,0.1,1,10,100],
    'feature_selection__percentile': np.arange(1, 20)
}

GSCV_Lasso = GridSearchCV(pipe_Lasso, param_grid_Lasso, cv=5, scoring='neg_mean_squared_error')
GSCV_Lasso2 = GridSearchCV(pipe_Lasso2, param_grid_Lasso2, cv=5, scoring='neg_mean_squared_error')
GSCV_Lasso3 = GridSearchCV(pipe_Lasso3, param_grid_Lasso3, cv=5, scoring='neg_mean_squared_error')
GSCV_Lasso4 = GridSearchCV(pipe_Lasso4, param_grid_Lasso4, cv=5, scoring='neg_mean_squared_error')

GSCV_Lasso.fit(x_train_enc, y_train_price)
GSCV_Lasso2.fit(x_train_enc, y_train_price)
GSCV_Lasso3.fit(x_train_enc, y_train_price)
GSCV_Lasso4.fit(x_train_enc, y_train_price)

print("Best model:{}".format(GSCV_Lasso.best_estimator_))
print("Lasso best paramaters:{}".format(GSCV_Lasso.best_params_))
print("Koefisien/bobot:{}".format(GSCV_Lasso.best_estimator_.named_steps['reg'].coef_))
print("Intercept/bias:{}".format(GSCV_Lasso.best_estimator_.named_steps['reg'].intercept_))

print("\n")

print("Best model:{}".format(GSCV_Lasso2.best_estimator_))
print("Lasso best paramaters:{}".format(GSCV_Lasso2.best_params_))
print("Koefisien/bobot:{}".format(GSCV_Lasso2.best_estimator_.named_steps['reg'].coef_))
print("Intercept/bias:{}".format(GSCV_Lasso2.best_estimator_.named_steps['reg'].intercept_))

print("\n")

print("Best model:{}".format(GSCV_Lasso3.best_estimator_))
print("Lasso best paramaters:{}".format(GSCV_Lasso3.best_params_))
print("Koefisien/bobot:{}".format(GSCV_Lasso3.best_estimator_.named_steps['reg'].coef_))
print("Intercept/bias:{}".format(GSCV_Lasso3.best_estimator_.named_steps['reg'].intercept_))

print("\n")

print("Best model:{}".format(GSCV_Lasso4.best_estimator_))
print("Lasso best paramaters:{}".format(GSCV_Lasso4.best_params_))
print("Koefisien/bobot:{}".format(GSCV_Lasso4.best_estimator_.named_steps['reg'].coef_))
print("Intercept/bias:{}".format(GSCV_Lasso4.best_estimator_.named_steps['reg'].intercept_))

Lasso_predict = GSCV_Lasso.predict(x_test_enc)
Lasso_predict2 = GSCV_Lasso2.predict(x_test_enc)
Lasso_predict3 = GSCV_Lasso3.predict(x_test_enc)
Lasso_predict4 = GSCV_Lasso4.predict(x_test_enc)

mse_Lasso = mean_squared_error(y_test_price, Lasso_predict)
mae_Lasso = mean_absolute_error(y_test_price, Lasso_predict)

mse_Lasso2 = mean_squared_error(y_test_price, Lasso_predict2)
mae_Lasso2 = mean_absolute_error(y_test_price, Lasso_predict2)

mse_Lasso3 = mean_squared_error(y_test_price, Lasso_predict3)
mae_Lasso3 = mean_absolute_error(y_test_price, Lasso_predict3)

mse_Lasso4 = mean_squared_error(y_test_price, Lasso_predict4)
mae_Lasso4 = mean_absolute_error(y_test_price, Lasso_predict4)

print("Lasso Mean Squared Error (MSE): {}".format(mse_Lasso))
print("Lasso Mean Absolute Error (MAE): {}".format(mae_Lasso))
print("Lasso Root Mean Squared Error: {}".format(np.sqrt(mse_Lasso)))

print("\n")

print("Lasso Mean Squared Error (MSE): {}".format(mse_Lasso2))
print("Lasso Mean Absolute Error (MAE): {}".format(mae_Lasso2))
print("Lasso Root Mean Squared Error: {}".format(np.sqrt(mse_Lasso2)))

print("\n")

print("Lasso Mean Squared Error (MSE): {}".format(mse_Lasso3))
print("Lasso Mean Absolute Error (MAE): {}".format(mae_Lasso3))
print("Lasso Root Mean Squared Error: {}".format(np.sqrt(mse_Lasso3)))

print("\n")

print("Lasso Mean Squared Error (MSE): {}".format(mse_Lasso4))
print("Lasso Mean Absolute Error (MAE): {}".format(mae_Lasso4))
print("Lasso Root Mean Squared Error: {}".format(np.sqrt(mse_Lasso4)))
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.feature_selection import SelectKBest, SelectPercentile, f_regression
from sklearn.metrics import mean_absolute_error, mean_squared_error

pipe_RFR = Pipeline(steps=[
            ('scale', StandardScaler()),
            ('feature_selection', SelectKBest(score_func=f_regression)),
            ('reg', RandomForestRegressor(random_state=40))            
            ])
pipe_RFR2 = Pipeline(steps=[
            ('scale', StandardScaler()),
            ('feature_selection', SelectPercentile(score_func=f_regression)),
            ('reg', RandomForestRegressor(random_state=40))            
            ])
pipe_RFR3 = Pipeline(steps=[
            ('scale', MinMaxScaler()),
            ('feature_selection', SelectKBest(score_func=f_regression)),
            ('reg', RandomForestRegressor(random_state=40))            
            ])
pipe_RFR4 = Pipeline(steps=[
            ('scale', MinMaxScaler()),
            ('feature_selection', SelectPercentile(score_func=f_regression)),
            ('reg', RandomForestRegressor(random_state=40))            
            ])

param_grid_RFR = {
    'feature_selection__k': np.arange(1, 20),
    'reg__n_estimators': [100, 200, 300],
    'reg__max_depth': [None, 10, 20, 30],
    'reg__min_samples_split': [2, 5, 10]
}
param_grid_RFR2 = {
    'feature_selection__percentile': np.arange(1, 20),
    'reg__n_estimators': [100, 200, 300],
    'reg__max_depth': [None, 10, 20, 30],
    'reg__min_samples_split': [2, 5, 10]
}
param_grid_RFR3 = {
    'feature_selection__k': np.arange(1, 20),
    'reg__n_estimators': [100, 200, 300],
    'reg__max_depth': [None, 10, 20, 30],
    'reg__min_samples_split': [2, 5, 10]
}
param_grid_RFR4 = {
    'feature_selection__percentile': np.arange(1, 20),
    'reg__n_estimators': [100, 200, 300],
    'reg__max_depth': [None, 10, 20, 30],
    'reg__min_samples_split': [2, 5, 10]
}

GSCV_RFR = GridSearchCV(pipe_RFR, param_grid_RFR, cv=5, scoring='neg_mean_squared_error', n_jobs=-1)
GSCV_RFR2 = GridSearchCV(pipe_RFR2, param_grid_RFR2, cv=5, scoring='neg_mean_squared_error', n_jobs=-1)
GSCV_RFR3 = GridSearchCV(pipe_RFR3, param_grid_RFR3, cv=5, scoring='neg_mean_squared_error', n_jobs=-1)
GSCV_RFR4 = GridSearchCV(pipe_RFR4, param_grid_RFR4, cv=5, scoring='neg_mean_squared_error', n_jobs=-1)


GSCV_RFR.fit(x_train_enc, y_train_price)
GSCV_RFR2.fit(x_train_enc, y_train_price)
GSCV_RFR3.fit(x_train_enc, y_train_price)
GSCV_RFR4.fit(x_train_enc, y_train_price)

for i, model in enumerate([GSCV_RFR, GSCV_RFR2, GSCV_RFR3, GSCV_RFR4], 1):
    print(f"\nModel {i} Results:")
    print("Best model:", model.best_estimator_)
    print("Best parameters:", model.best_params_)
    print("Feature importances:", model.best_estimator_.named_steps['reg'].feature_importances_)

    predictions = model.predict(x_test_enc)
    
    mse = mean_squared_error(y_test_price, predictions)
    mae = mean_absolute_error(y_test_price, predictions)
    rmse = np.sqrt(mse)
    
    print("\nModel Performance Metrics:")
    print(f"Mean Squared Error (MSE): {mse:.4f}")
    print(f"Mean Absolute Error (MAE): {mae:.4f}")
    print(f"Root Mean Squared Error (RMSE): {rmse:.4f}")
    print("-" * 50)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Membuat DataFrame perbandingan
def create_comparison_df(y_true, pred_rf, pred_lasso, model1_name="Random Forest", model2_name="Lasso"):
    comparison_df = pd.DataFrame({
        'Actual_Price': y_true,
        f'{model1_name}_Prediction': pred_rf,
        f'{model2_name}_Prediction': pred_lasso,
        f'{model1_name}_Difference': pred_rf - y_true,
        f'{model2_name}_Difference': pred_lasso - y_true
    })
    
    # Menambahkan persentase error
    comparison_df[f'{model1_name}_Error_Percentage'] = (abs(comparison_df[f'{model1_name}_Difference']) / comparison_df['Actual_Price'] * 100).round(2)
    comparison_df[f'{model2_name}_Error_Percentage'] = (abs(comparison_df[f'{model2_name}_Difference']) / comparison_df['Actual_Price'] * 100).round(2)
    
    return comparison_df
```
