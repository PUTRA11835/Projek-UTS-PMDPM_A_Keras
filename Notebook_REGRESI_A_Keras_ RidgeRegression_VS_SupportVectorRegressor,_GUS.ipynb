```python
import pandas as pd
import numpy as np

df_properti = pd.read_csv('Dataset UTS_Gasal 2425.csv')
df_properti.head(10)
```

```python
df_properti2 = df_properti.drop(['category'], axis=1)
df_properti2.head()
```

```python
print("Sebelum Pengecekkan data duplikat", df_properti2.shape)
df_properti3 = df_properti2.drop_duplicates(keep='last')
print("Sesudah Pengecekkan data duplikat", df_properti3.shape)
```

```python
from sklearn.model_selection import train_test_split

x_regress = df_properti3.drop('price', axis=1)
y_regress = df_properti3.price

x_train_price, x_test_price, y_train_price, y_test_price = train_test_split(x_regress, y_regress, test_size=0.25, random_state=40)
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer

kolom_kategori=['hasyard','haspool','isnewbuilt','hasstormprotector','hasstorageroom']

transform= make_column_transformer(
    (OneHotEncoder(),kolom_kategori),remainder='passthrough'
)
```

```python
x_train_enc=transform.fit_transform(x_train_price)
x_test_enc=transform.transform(x_test_price)

df_train_enc=pd.DataFrame(x_train_enc,columns=transform.get_feature_names_out())
df_test_enc=pd.DataFrame(x_test_enc,columns=transform.get_feature_names_out())

df_train_enc.head(10)
df_test_enc.head(10)
```

```python
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.feature_selection import SelectKBest, SelectPercentile, f_regression
from sklearn.metrics import mean_absolute_error, mean_squared_error

pipe_Ridge = Pipeline(steps=[
            ('scale', StandardScaler()),
            ('feature_selection', SelectKBest(score_func=f_regression)),
            ('reg', Ridge())
            ])

pipe_Ridge2 = Pipeline(steps=[
            ('scale', MinMaxScaler()),
            ('feature_selection', SelectPercentile(score_func=f_regression)),
            ('reg', Ridge())
            ])

pipe_Ridge3 = Pipeline(steps=[
            ('scale', StandardScaler()),
            ('feature_selection', SelectPercentile(score_func=f_regression)),
            ('reg', Ridge())
            ])

pipe_Ridge4 = Pipeline(steps=[
            ('scale', MinMaxScaler()),
            ('feature_selection', SelectKBest(score_func=f_regression)),
            ('reg', Ridge())
            ])

param_grid_Ridge = {
    'reg__alpha': [0.01,0.1,1,10,100],
    'feature_selection__k': np.arange(1,20)
}

param_grid_Ridge2 = {
    'reg__alpha': [0.01,0.1,1,10,100],
    'feature_selection__percentile': np.arange(1,20)
}

param_grid_Ridge3 = {
    'reg__alpha': [0.01,0.1,1,10,100],
    'feature_selection__percentile': np.arange(1,20)
}

param_grid_Ridge4 = {
    'reg__alpha': [0.01,0.1,1,10,100],
    'feature_selection__k': np.arange(1,20)
}

GSCV_RR = GridSearchCV(pipe_Ridge, param_grid_Ridge, cv=5,
                            scoring='neg_mean_squared_error', error_score='raise')

GSCV_RR2 = GridSearchCV(pipe_Ridge2, param_grid_Ridge2, cv=5,
                            scoring='neg_mean_squared_error', error_score='raise')

GSCV_RR3 = GridSearchCV(pipe_Ridge3, param_grid_Ridge3, cv=5,
                            scoring='neg_mean_squared_error', error_score='raise')

GSCV_RR4 = GridSearchCV(pipe_Ridge4, param_grid_Ridge4, cv=5,
                            scoring='neg_mean_squared_error', error_score='raise')

GSCV_RR.fit(x_train_enc, y_train_price)
GSCV_RR2.fit(x_train_enc, y_train_price)
GSCV_RR3.fit(x_train_enc, y_train_price)
GSCV_RR4.fit(x_train_enc, y_train_price)

print("Best mode:{}".format(GSCV_RR.best_estimator_))
print("Ridge best parameters:{}".format(GSCV_RR.best_params_))

print("Best mode:{}".format(GSCV_RR2.best_estimator_))
print("Ridge best parameters:{}".format(GSCV_RR2.best_params_))

print("Best mode:{}".format(GSCV_RR3.best_estimator_))
print("Ridge best parameters:{}".format(GSCV_RR3.best_params_))

print("Best mode:{}".format(GSCV_RR4.best_estimator_))
print("Ridge best parameters:{}".format(GSCV_RR4.best_params_))

print("Koefisien/bobot:{}".format(GSCV_RR.best_estimator_.named_steps['reg'].coef_))
print("Intercept/bias:{}".format(GSCV_RR.best_estimator_.named_steps['reg'].intercept_))

print("Koefisien/bobot:{}".format(GSCV_RR2.best_estimator_.named_steps['reg'].coef_))
print("Intercept/bias:{}".format(GSCV_RR2.best_estimator_.named_steps['reg'].intercept_))

print("Koefisien/bobot:{}".format(GSCV_RR3.best_estimator_.named_steps['reg'].coef_))
print("Intercept/bias:{}".format(GSCV_RR3.best_estimator_.named_steps['reg'].intercept_))

print("Koefisien/bobot:{}".format(GSCV_RR4.best_estimator_.named_steps['reg'].coef_))
print("Intercept/bias:{}".format(GSCV_RR4.best_estimator_.named_steps['reg'].intercept_))

Ridge_predict = GSCV_RR.predict(x_test_enc)
Ridge_predict2 = GSCV_RR2.predict(x_test_enc)
Ridge_predict3 = GSCV_RR3.predict(x_test_enc)
Ridge_predict4 = GSCV_RR4.predict(x_test_enc)

mse_Ridge = mean_squared_error(y_test_price, Ridge_predict)
mae_Ridge = mean_absolute_error(y_test_price, Ridge_predict)

mse_Ridge2 = mean_squared_error(y_test_price, Ridge_predict2)
mae_Ridge2 = mean_absolute_error(y_test_price, Ridge_predict2)

mse_Ridge3 = mean_squared_error(y_test_price, Ridge_predict3)
mae_Ridge3 = mean_absolute_error(y_test_price, Ridge_predict3)

mse_Ridge4 = mean_squared_error(y_test_price, Ridge_predict4)
mae_Ridge4 = mean_absolute_error(y_test_price, Ridge_predict4)

print("Ridge Mean Squared Error (MSE): {}".format(mse_Ridge))
print("Ridge Mean Absolute Error (MAE): {}".format(mae_Ridge))
print("Ridge Root Mean Squared Error: {}".format(np.sqrt(mse_Ridge)))

print("Ridge Mean Squared Error (MSE): {}".format(mse_Ridge2))
print("Ridge Mean Absolute Error (MAE): {}".format(mae_Ridge2))
print("Ridge Root Mean Squared Error: {}".format(np.sqrt(mse_Ridge2)))

print("Ridge Mean Squared Error (MSE): {}".format(mse_Ridge3))
print("Ridge Mean Absolute Error (MAE): {}".format(mae_Ridge3))
print("Ridge Root Mean Squared Error: {}".format(np.sqrt(mse_Ridge3)))

print("Ridge Mean Squared Error (MSE): {}".format(mse_Ridge4))
print("Ridge Mean Absolute Error (MAE): {}".format(mae_Ridge4))
print("Ridge Root Mean Squared Error: {}".format(np.sqrt(mse_Ridge4)))
```

```python
from sklearn.svm import SVR
from sklearn.model_selection import GridSearchCV
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.feature_selection import SelectKBest, SelectPercentile, f_regression
from sklearn.metrics import mean_absolute_error, mean_squared_error

pipe_SVR = Pipeline(steps=[
            ('scale', StandardScaler()),
            ('feature_selection', SelectKBest(score_func=f_regression)),
            ('reg', SVR(kernel='linear'))
            ])

pipe_SVR2 = Pipeline(steps=[
            ('scale', MinMaxScaler()),
            ('feature_selection', SelectPercentile(score_func=f_regression)),
            ('reg', SVR(kernel='linear'))
            ])

pipe_SVR3 = Pipeline(steps=[
            ('scale', StandardScaler()),
            ('feature_selection', SelectPercentile(score_func=f_regression)),
            ('reg', SVR(kernel='linear'))
            ])

pipe_SVR4 = Pipeline(steps=[
            ('scale', MinMaxScaler()),
            ('feature_selection', SelectKBest(score_func=f_regression)),
            ('reg', SVR(kernel='linear'))
            ])

param_grid_SVR = {
    'reg__C': [0.01, 0.1, 1, 10, 100],
    'reg__epsilon': [0.1, 0.2, 0.5, 1],
    'feature_selection__k': np.arange(1,20)
}

param_grid_SVR2 = {
    'reg__C': [0.01, 0.1, 1, 10, 100],
    'reg__epsilon': [0.1, 0.2, 0.5, 1],
    'feature_selection__percentile': np.arange(1,20)
}

param_grid_SVR3 = {
    'reg__C': [0.01, 0.1, 1, 10, 100],
    'reg__epsilon': [0.1, 0.2, 0.5, 1],
    'feature_selection__percentile': np.arange(1,20)
}

param_grid_SVR4 = {
    'reg__C': [0.01, 0.1, 1, 10, 100],
    'reg__epsilon': [0.1, 0.2, 0.5, 1],
    'feature_selection__k': np.arange(1,20)
}

GSCV_SVR = GridSearchCV(pipe_SVR,param_grid_SVR, cv=5, scoring='neg_mean_squared_error')
GSCV_SVR2 = GridSearchCV(pipe_SVR2,param_grid_SVR2, cv=5, scoring='neg_mean_squared_error')
GSCV_SVR3 = GridSearchCV(pipe_SVR3,param_grid_SVR3, cv=5, scoring='neg_mean_squared_error')
GSCV_SVR4 = GridSearchCV(pipe_SVR4,param_grid_SVR4, cv=5, scoring='neg_mean_squared_error')

GSCV_SVR.fit(x_train_enc, y_train_price)
GSCV_SVR2.fit(x_train_enc, y_train_price)
GSCV_SVR3.fit(x_train_enc, y_train_price)
GSCV_SVR4.fit(x_train_enc, y_train_price)

print("Best model:{}".format(GSCV_SVR.best_estimator_))
print("SVR best parameters:{}".format(GSCV_SVR.best_params_))

print("Best model:{}".format(GSCV_SVR2.best_estimator_))
print("SVR best parameters:{}".format(GSCV_SVR2.best_params_))

print("Best model:{}".format(GSCV_SVR3.best_estimator_))
print("SVR best parameters:{}".format(GSCV_SVR3.best_params_))

print("Best model:{}".format(GSCV_SVR3.best_estimator_))
print("SVR best parameters:{}".format(GSCV_SVR3.best_params_))

print("Koefisien/bobot:{}".format(GSCV_SVR4.best_estimator_.named_steps['reg'].coef_))
print("Intercept/bias:{}".format(GSCV_SVR4.best_estimator_.named_steps['reg'].intercept_))

SVR_predict = GSCV_SVR.predict(x_test_enc)
SVR_predict2 = GSCV_SVR2.predict(x_test_enc)
SVR_predict3 = GSCV_SVR3.predict(x_test_enc)
SVR_predict4 = GSCV_SVR4.predict(x_test_enc)

mse_SVR = mean_squared_error(y_test_price, SVR_predict)
mae_SVR = mean_absolute_error(y_test_price, SVR_predict)

mse_SVR2 = mean_squared_error(y_test_price, SVR_predict2)
mae_SVR2 = mean_absolute_error(y_test_price, SVR_predict2)

mse_SVR3 = mean_squared_error(y_test_price, SVR_predict3)
mae_SVR3 = mean_absolute_error(y_test_price, SVR_predict3)

mse_SVR4 = mean_squared_error(y_test_price, SVR_predict4)
mae_SVR4 = mean_absolute_error(y_test_price, SVR_predict4)

print("SVR Mean Squared Error (MSE): {}".format(mse_SVR))
print("SVR Mean Absolute Error (MAE): {}".format(mae_SVR))
print("SVR Root Mean Squared Error: {}".format(np.sqrt(mse_SVR)))

print("SVR Mean Squared Error (MSE): {}".format(mse_SVR2))
print("SVR Mean Absolute Error (MAE): {}".format(mae_SVR2))
print("SVR Root Mean Squared Error: {}".format(np.sqrt(mse_SVR2)))

print("SVR Mean Squared Error (MSE): {}".format(mse_SVR3))
print("SVR Mean Absolute Error (MAE): {}".format(mae_SVR3))
print("SVR Root Mean Squared Error: {}".format(np.sqrt(mse_SVR3)))

print("SVR Mean Squared Error (MSE): {}".format(mse_SVR4))
print("SVR Mean Absolute Error (MAE): {}".format(mae_SVR4))
print("SVR Root Mean Squared Error: {}".format(np.sqrt(mse_SVR4)))
```
